## SMOTE(Synthetic Minority Oversampling Technique):-
It is a technique used in machine learning to address imbalanced datasets where the minority class has significantly fewer instances than the majority class. SMOTE involves generating synthetic instances of the minority class by interpolating between existing instances.

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=1,weights=[0.90],random_state=12)
```

```python
import pandas as pd
df1 = pd.DataFrame(X,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
```

```python
df1
```

```python
df2
```

```python
df = pd.concat([df1,df2],axis=1)
df
```

```python
df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['f1'],df['f2'],c=df['target'])
```

```python
from imblearn.over_sampling import SMOTE
#transform the dataset
oversample = SMOTE()
X,y = oversample.fit_resample(df[['f1','f2']],df['target'])
```
